## Wprowadzenie

Równanie Schrödingera jest fundamentalnym równaniem w mechanice kwantowej, opisującym ewolucję funkcji falowej cząstki. Jest to liniowe równanie różniczkowe cząstkowe rzędu pierwszego względem czasu i rzędu drugiego względem przestrzeni. Równanie opisuje, jak stan kwantowy układu zmienia się w czasie, co pozwala na przewidywanie zachowania cząstek na poziomie kwantowym.

## Model Matematyczny
Równanie Schrödingera dane jest wzorem:

$$i\hbar \frac{\partial}{\partial t} \Psi(x,t) = \left[ -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + V(x,t) \right] \Psi(x,t)$$
$$i\hbar \frac{\partial}{\partial t} \Psi(x,t) =  \hat{H} \Psi(x,t)$$

gdzie:
- $i$ to jednostka urojona (dzięki niej rozwiązanie ma charakter falowy),
- $\hbar$ to zredukowana stała Plancka,
- $\Psi(x,t)$ to funkcja falowa,
- $m$ to masa cząstki,
- $V(x,t)$ to potencjał działający na cząstkę,
- $\hat{H}$ to operator Hamiltona, reprezentujący całkowitą energię układu.

## Funkcja falowa

W mechanice kwantowej nie znamy dokładnej trajektorii cząstki, dlatego używamy funkcji falowej $\Psi(x,t)$, która zawiera informacje o stanie kwantowym cząstki. Funkcja falowa jest zespolona, a jej moduł kwadratowy $|\Psi(x,t)|^2$ daje gęstość prawdopodobieństwa znalezienia cząstki w określonym miejscu i czasie. Oznacza to, że im większa wartość $|\Psi(x,t)|^2$, tym większe prawdopodobieństwo znalezienia cząstki w danym obszarze przestrzeni. Funkcja falowa zawiera informacje o wszystkich możliwych stanach układu i jest podstawą do obliczania energii czy pędu.

## Potencjał

Potencjał $V(x, t)$ reprezentuje siły działające na cząstkę. W mechanice kwatnowej reprezentuje on energię potencjalną układu. Potencjał może być funkcją zarówno przestrzeni, jak i czasu, co pozwala na modelowanie różnych sytuacji fizycznych. Może on przyjmować różne formy, takie jak potencjał studni, potencjał barierowy czy potencjał harmoniczny. W zależności od kształtu potencjału, funkcja falowa będzie miała różne właściwości i będzie opisywać różne stany kwantowe. 

## Założenia modelu matematycznego

Zakładamy, że:
- potencjał $V(x,t)$ jest stały w czasie (środowisko jest stacjonarne), czyli $V(x,t) = V(x)$,
- funkcja falowa $\Psi(x,t)$ może być rozdzielona na część zależną od czasu i część zależną od przestrzeni, czyli $\Psi(x,t) = \Psi(x) \cdot \phi(t)$,
- warunki brzegowe oraz początkowe są dobrze określone, oś x ograniczam do przedziału $[0, L]$ i zakładam, że funkcja falowa jest zerowa na brzegach tego przedziału (warunki Dirichleta),
- unitarność i ciągłość funkcji falowej, zatem $\int_0^L |\Psi(x,t)|^2 dx = 1$


## Przekształcenie modelu matematycznego

$$i\hbar \frac{\partial}{\partial t} \Psi(x,t) = \left[ -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + V(x) \right] \Psi(x,t)$$

$$\iff i\hbar \frac{\partial}{\partial t} [\Psi(x) \cdot \phi(t)] = \left[ -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + V(x) \right] [\Psi(x) \cdot \phi(t)]$$

$$\iff i\hbar \cdot \Psi(x) \frac{\partial \phi(t)}{\partial t} = \left[ -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi(x)}{\partial x^2} + V(x)\Psi(x) \right] \phi(t) \quad \Bigg| : \big(\Psi(x) \cdot \phi(t)\big)$$

$$\iff i\hbar \frac{1}{\phi(t)} \frac{\partial \phi(t)}{\partial t} = \frac{1}{\Psi(x)} \left[ -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi(x)}{\partial x^2} + V(x)\Psi(x) \right]$$

Zauważmy, że lewa strona równania zależy tylko od czasu, a prawa strona zależy tylko od przestrzeni. Oznacza to, że obie strony muszą być równe tej samej stałej, którą oznaczymy jako $E$ (energia układu):

$$
\begin{cases} 
\displaystyle i\hbar \frac{1}{\phi(t)} \frac{d\phi(t)}{dt} = E \\[15pt]
\displaystyle \frac{1}{\Psi(x)} \left[ -\frac{\hbar^2}{2m} \frac{d^2\Psi(x)}{dx^2} + V(x)\Psi(x) \right] = E 
\end{cases}
$$

Stąd otrzymujemy równanie, na którym się skupimy:
$$\frac{1}{\Psi(x)} \left[ -\frac{\hbar^2}{2m} \frac{d^2}{dx^2} + V(x) \right] = E \quad \Bigg| \cdot \big(\Psi(x)\big)$$

$$\iff [-\frac{\hbar^2}{2m} \frac{d^2 \Psi(x)}{dx^2} + V(x)]\Psi(x) = E \cdot \Psi(x)$$



Ostateczne otrzymujemy równanie Schrödingera w postaci stacjonarnej: $\hat{H} \Psi(x) = E \Psi(x)$
Mając powyższe równanie szukamy wartości własnych $E$ (reprezentujących dozwolone energie) i wektorów własnych $\Psi(x)$ operatora Hamiltona. To pozwoli nam zrozumieć stany energetyczne układu kwantowego.

## Przykład analityczny - potencjał studni
Najprostszym przypadkiem jest potencjał studni, gdzie $$V(x) = \begin{cases} 0 & \text{dla } 0 < x < L \\ \infty & \text{dla } x \le 0 \text{ oraz } x \ge L \end{cases}$$ Oznacza to, że cząstka jest ograniczona do poruszania się w przedziale $[0, L]$ i nie może przekroczyć tych granic. 
Wewnątrz studni potencjał jest równy zero, więc równanie Schrödingera upraszcza się do:
$$-\frac{\hbar^2}{2m} \frac{d^2 \Psi(x)}{dx^2} = E \cdot \Psi(x) \quad \Bigg| \cdot \big(-\frac{2m}{\hbar^2})$$
$$\frac{d^2\Psi(x)}{dx^2} + \frac{2mE}{\hbar^2}\Psi(x) = 0$$
Rozwiązaniem tego równania jest funkcja sinusoidalna:
$$\Psi(x) = A \sin(kx) + B \cos(kx)$$
gdzie $k = \sqrt{\frac{2mE}{\hbar^2}}$. 


Z warunków brzegowych mamy:
$$\Psi(0) = 0 \implies B = 0$$
$$\Psi(L) = 0 \implies A \sin(kL) = 0 \implies kL = n\pi, \quad n = 1, 2, 3, ...$$
Stąd otrzymujemy rozwiązanie:
$$\Psi_n(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)$$
oraz odpowiadające im energie:
$$E_n = \frac{n^2 \pi^2 \hbar^2}{2mL^2}$$


```julia
using Plots

L = 1       
Nx = 1000        
x = range(0, L, length=Nx) 

psi(x, n, L) = sqrt(2 / L) * sin(n * pi * x / L)


psi_n1 = [psi(xi, 1, L) for xi in x]
psi_n2 = [psi(xi, 2, L) for xi in x]

dens_n1 = psi_n1 .^ 2
dens_n2 = psi_n2 .^ 2


p1 = plot(x, psi_n1, label="n = 1", linewidth=2.5, color=:blue,
          title=raw"Funkcja falowa $\psi(x)$", ylabel=raw"$\psi(x)$", legend=:topright)
plot!(p1, x, psi_n2, label="n = 2", linewidth=2.5, color=:red)
hline!(p1, [0], color=:black, linestyle=:dash, label="") 

p2 = plot(x, dens_n1, label="n = 1", linewidth=2.5, color=:blue,
          title=raw"Gęstość prawdopodobieństwa $|\psi(x)|^2$", xlabel=raw"$x$", ylabel=raw"$|\psi(x)|^2$", legend=:topright)
plot!(p2, x, dens_n2, label="n = 2", linewidth=2.5, color=:red)


plt = plot(p1, p2, layout=(2, 1), size=(700, 600))

display(plt)
```

## Założenie A — Siatka musi być wystarczająco gęsta

$$\Delta x = \frac{L}{N-1}, \qquad N \gg 10n$$

**Co się psuje:** przy za małym $N$ siatka nie „widzi" maksimów funkcji — energia błędna, $\Psi$ kancista.

#### Testujemy stan n=5 przy różnych N — od katastrofalnego do dobrego

```julia
using Plots
using Plots.Measures

L = 1 
psi_analytic(x, n) = sqrt(2/L) * sin(n * π * x / L)
E_analytic(n)      = (n^2 * π^2) / (2 * L^2)

n_test = 5
N_values = [8, 15, 30, 200]
col = [:red, :orange, :dodgerblue, :green]
x_fine = range(0, L, length=1000)

plots_A = []
for (i, N) in enumerate(N_values)
    x_grid = range(0, L, length=N)
    Δx = L / (N - 1)
    err_pct = (n_test^2 * π^2 * Δx^2) / 12 * 100

    tytul_suwaka = "\nN=$N  błąd E=$(round(err_pct, sigdigits=2))%"

    p = plot(x_fine, psi_analytic.(x_fine, n_test),
        label="analityczna", color=:black, lw=1.2, ls=:dash, alpha=0.5,
        title=tytul_suwaka,
        titlefontsize=10, 
        xlabel="x", 
        ylabel="Ψ₅(x)", 
        legendfontsize=7)
        
    plot!(p, collect(x_grid), psi_analytic.(x_grid, n_test),
        label="siatka", color=col[i], lw=2, markershape=:circle, markersize=3)
    push!(plots_A, p)
end

plot(plots_A..., 
     layout=(2,2), 
     size=(800, 580), 
     left_margin=5mm,
     right_margin=5mm,
     plot_title="Wpływ gęstości siatki N na błąd numeryczny energii (dla stanu n=5)",
     plot_titlefontsize=13)
```

**Wniosek:** Dla $N = 8$ błąd energii wynosi kilkanaście procent — wynik jest fizycznie bezużyteczny.  
Błąd maleje jak $O(\Delta x^2)$, co jest charakterystyczne dla metody FDM którą zaraz opiszemy.  
W pakiecie Julia zaimplementujemy automatyczne ostrzeżenie gdy $N < 20n$.

## Założenie B — Normalizacja: norma $\ell^2$ ≠ norma całkowa

Fizyczny warunek $\int_0^L |\Psi|^2\,dx = 1$ zastępujemy dyskretnie regułą trapezów:

$$\sum_{i=1}^{N} |\Psi_i|^2 \,\Delta x \approx 1$$

**Co może pójść źle:** `eigen()` w Julii zwraca wektory znormalizowane w sensie $\ell^2$ (suma kwadratów = 1), **nie** w sensie całkowym. Różnica wynosi czynnik $\sim 1/\Delta x$ — dla $N = 1000$ to błąd rzędu tysiąckrotnego.

```julia
using LinearAlgebra, Printf

trapezoid_norm(psi, dx) =
    (0.5*psi[1]^2 + sum(psi[2:end-1].^2) + 0.5*psi[end]^2) * dx

function build_H(N)
    Δx = L / (N - 1)
    t  = 1.0 / (2 * Δx^2)
    SymTridiagonal(fill(2t, N-2), fill(-t, N-3))
end

H    = build_H(80)
vals, vecs = eigen(H)
Δx2  = L / 79

println("n  | norma l2 (solver) | norma całkowa      | po renorm.")
println("-"^62)
for n in 1:4
    psi_in   = vecs[:, n]
    norm_l2  = sum(psi_in.^2)
    psi_full = [0.0; psi_in; 0.0]
    norm_tr  = trapezoid_norm(psi_full, Δx2)
    psi_full ./= sqrt(norm_tr)
    err      = abs(1.0 - trapezoid_norm(psi_full, Δx2))
    @printf("%-3d| %-19.4f | %-20.10f | %.2e\n", n, norm_l2, norm_tr, err)
end
println("\nNorma l2 ≠ 1 — różni się czynnikiem ~1/Δx = $(round(1/Δx2, digits=1)).")
```

**Wniosek:** Błędna implementacja daje $\Psi(0) \neq 0$ — złamanie podstawowego warunku fizycznego.  
Poprawna implementacja: macierz wymiaru $(N-2) \times (N-2)$, wartości $\Psi = 0$ na brzegach doklejamy po fakcie.

## Algorytm 1 — Metoda Różnic Skończonych (FDM)

Przybliżamy drugą pochodną centralnym ilorazem różnicowym:

$$\frac{d^2\Psi_i}{dx^2} \approx \frac{\Psi_{i+1} - 2\Psi_i + \Psi_{i-1}}{\Delta x^2} + O(\Delta x^2)$$

Równanie $\hat{H}\Psi = E\Psi$ staje się **macierzowym zagadnieniem własnym** $H\mathbf{\Psi} = E\mathbf{\Psi}$, gdzie $H$ jest macierzą trójprzekątniową:

$$H_{ii} = \frac{\hbar^2}{m\Delta x^2} + V(x_i), \qquad H_{i,i\pm1} = -\frac{\hbar^2}{2m\Delta x^2}$$

**Złożoność:**
- Solver gęsty (pełny QR): $O(N^3)$ — dla $N = 10^4$ to $10^{12}$ operacji
- `SymTridiagonal` + Krylov/Lanczos: $O(N \cdot k)$, gdzie $k$ = liczba szukanych stanów

```julia
# Pełna implementacja FDM z dowolnym potencjałem

function solve_FDM(N, V_func; n_states=5)
    x_full  = range(0, L, length=N)
    x_inner = x_full[2:end-1]
    Δx      = step(x_full)
    t       = 1.0 / (2 * Δx^2)        # ħ = m = 1

    d_main = [2t + V_func(x_inner[i]) for i in 1:N-2]
    d_off  = fill(-t, N-3)
    H      = SymTridiagonal(d_main, d_off)

    vals, vecs = eigen(H)

    wavefns = zeros(N, n_states)
    for k in 1:n_states
        psi_full = [0.0; vecs[:,k]; 0.0]
        psi_full ./= sqrt(trapezoid_norm(psi_full, Δx))
        psi_full .*= sign(psi_full[max(1, round(Int, N/4))])
        wavefns[:,k] = psi_full
    end
    return vals[1:n_states], wavefns, collect(x_full)
end

V_zero(x) = 0.0
energies, wavefns, x_grid = solve_FDM(300, V_zero, n_states=4)

x_fine = range(0, L, length=1000)
cols4  = [:dodgerblue, :crimson, :green, :purple]

p1 = plot(title="Funkcje falowe Ψₙ(x)", xlabel="x", ylabel="Ψ(x)")
p2 = plot(title="Gęstości prawdopodobieństwa |Ψₙ(x)|²", xlabel="x", ylabel="|Ψ|²")

for n in 1:4
    lab = "n=$n  E=$(round(energies[n], digits=3))"
    plot!(p1, x_grid, wavefns[:,n], label=lab, color=cols4[n], lw=2)
    plot!(p2, x_grid, wavefns[:,n].^2, label=lab, color=cols4[n], lw=2)
end
hline!(p1, [0], color=:black, ls=:dash, lw=0.8, label="")
plot(p1, p2, layout=(2,1), size=(700,500))
```

## Algorytm 2 — Metoda Numerowa: $O(\Delta x^4)$ zamiast $O(\Delta x^2)$

Dla równań postaci $\Psi'' = f(x)\Psi$ istnieje dedykowana metoda uwzględniająca wyższe wyrazy rozwinięcia Taylora:

$$\Psi_{i+1}\!\left(1 + \tfrac{\Delta x^2}{12}f_{i+1}\right) = 2\Psi_i\!\left(1 - \tfrac{5\Delta x^2}{12}f_i\right) - \Psi_{i-1}\!\left(1 + \tfrac{\Delta x^2}{12}f_{i-1}\right)$$

Energię własną szukamy bisekcją: propagujemy $\Psi$ od lewego brzegu dla danego $E$ i szukamy $E$ takich że $\Psi(L) = 0$.

| Metoda | Rząd | $N$ potrzebne do błędu $10^{-6}$ |
|--------|------|----------------------------------|
| FDM    | $O(\Delta x^2)$ | $N \approx 1000$ |
| Numerov | $O(\Delta x^4)$ | $N \approx 30$ |

**Alternatywy:** metoda elementów skończonych (FEM), metoda spektralna, metoda strzału.

```julia
function numerov_shoot(E, N)
    x   = range(0, L, length=N)
    Δx  = step(x)
    f   = -2.0 * E          # V=0, ħ=m=1, stałe

    w_side   = 1.0 + Δx^2/12 * f
    w_center = 1.0 - 5*Δx^2/12 * f

    psi    = zeros(N)
    psi[1] = 0.0
    psi[2] = Δx              # krok startowy — bez sin, bez małego 1e-3

    for i in 2:N-1
        psi[i+1] = (2*psi[i]*w_center - psi[i-1]*w_side) / w_side
    end
    return psi[end]
end
```

```julia
function solve_Numerov(N, V_func; n_states=5)
    x    = range(0, L, length=N)
    Δx   = step(x)
    x_in = collect(x[2:end-1])
    M    = N - 2
    t    = 1.0 / (2 * Δx^2)   # ħ=m=1

    # Macierz H (jak FDM)
    A = SymTridiagonal(
        [2t + V_func(x_in[i]) for i in 1:M],
        fill(-t, M-1)
    )

    # Macierz masy B — poprawka Numerova: B*ψ po prawej stronie
    # B_ii = 10/12, B_{i,i±1} = 1/12
    B = SymTridiagonal(
        fill(10/12, M),
        fill(1/12, M-1)
    )

    # Rozwiązujemy A*v = E*B*v  →  B^{-1}*A*v = E*v
    # B jest SPD więc można: Cholesky
    C   = cholesky(Matrix(B))
    Ĥ   = C.L \ Matrix(A) / C.U        # C.L⁻¹ A C.U⁻¹

    vals, vecs_hat = eigen(Symmetric(Ĥ))
    vecs = C.U \ vecs_hat               # powrót do oryginalnej bazy

    wavefns = zeros(N, n_states)
    for k in 1:n_states
        psi_full = [0.0; vecs[:,k]; 0.0]
        psi_full ./= sqrt(trapezoid_norm(psi_full, Δx))
        psi_full .*= sign(psi_full[max(1, round(Int, N/4))])
        wavefns[:,k] = psi_full
    end
    return vals[1:n_states], wavefns, collect(x)
end
```

```julia
N_vals = [8, 10, 15, 20, 30, 50, 100, 200]

err_fdm     = [abs(solve_FDM(N, V_zero, n_states=1)[1][1]     - E_analytic(1)) / E_analytic(1) for N in N_vals]
err_numerov = [abs(solve_Numerov(N, V_zero, n_states=1)[1][1] - E_analytic(1)) / E_analytic(1) for N in N_vals]

plot(N_vals, err_fdm,
    label="FDM — O(Δx²)", color=:crimson, lw=2.5,
    markershape=:circle, markersize=5,
    yscale=:log10, xscale=:log10,
    xlabel="N", ylabel="błąd względny E₁",
    title="FDM vs Numerov — zbieżność", legend=:bottomleft)
plot!(N_vals, err_numerov,
    label="Numerov — O(Δx⁴)", color=:green, lw=2.5,
    markershape=:diamond, markersize=5, ls=:dash)
```
